```
import numpy as np
```

```
f=open("/content/drive/MyDrive/nlp/dinos.txt")
```

```
names=f.read().lower()
chars=list(set(names))
data_size,vocab_size=len(names),len(chars)
print("data size %f, vocab_size %f"%(data_size,vocab_size))
f.close()
```

```
names
```

```
chars = sorted(chars)
print(chars)
```

```
char_to_ix={c:i for i,c in enumerate(chars)}
ix_to_char={i:c for i,c in enumerate(chars)}
```

```
print(char_to_ix)
print(ix_to_char)
```

```

def sample1(preds, temperature=1.0):
    # helper function to sample an index from a probability array
    preds = np.asarray(preds).astype("float64")
    preds = np.log(preds) / temperature
    exp_preds = np.exp(preds)
    preds = exp_preds / np.sum(exp_preds)
    probas = np.random.multinomial(1, preds, 1)
    return np.argmax(probas)
```

```
def sample(model):
  x=np.zeros((1,1,vocab_size))
  idx=-1
  gench = []
  counter=0
  eof_char=char_to_ix["\n"]

  while (idx!=eof_char and counter!=50):
      probs=model.predict(x)[0]
      idx = sample1(probs)
      gench.append(idx)
      x = np.zeros((1,1,vocab_size))
      x[:,:,idx] = 1
      counter +=1
      if (counter == 50):
          gench.append(eof_char)
  return gench
```

```
def print_name(sample_idx,ix_to_char):
  name=""
  for idx in sample_idx:
    name=name+ix_to_char[idx]
  return name
```

```
from tensorflow import keras
from tensorflow.keras import layers
```

```
model = keras.Sequential(
    [
        keras.Input(shape=(1, 27)),
        layers.LSTM(128,return_sequences=False),
        layers.Dense(vocab_size, activation="softmax"),
    ]
)
optimizer = keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss="categorical_crossentropy", optimizer=optimizer)
```

```
X=np.zeros((len(names),1,vocab_size))

for i,a in enumerate(names):
  idx=char_to_ix[a]
  X[i,:,idx]=1
y=X[1:,0,:]
```

```
print(X[:-1,:,:].shape,y.shape)
```

```
epoch=40
batch_=16
for e in range(epoch):
  model.fit(X[:-1,:,:],y,batch_size=batch_,epochs=1)
  for g in range(10):
    s=sample(model)
    print(print_name(s,ix_to_char))
```
